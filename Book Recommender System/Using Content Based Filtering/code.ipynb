```python
def clean_and_concatenate(row):
    # Convert lists to strings
    book_desc = ', '.join(row['book_desc']) if isinstance(row['book_desc'], list) else row['book_desc']
    genres = ', '.join(row['genres']) if isinstance(row['genres'], list) else row['genres']
    book_authors = ', '.join(row['book_authors']) if isinstance(row['book_authors'], list) else row['book_authors']
    book_format = ', '.join(row['book_format']) if isinstance(row['book_format'], list) else row['book_format']

    # Clean and concatenate the columns
    return book_desc + ', ' + genres + ', ' + book_authors + ', ' + book_format
```

```python
books['tags'] = books.apply(clean_and_concatenate, axis=1)
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))

    return " ".join(l)
```

```python
new_books['tags'] = new_books['tags'].apply(stems)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
new_books.iloc[0]['tags']
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(new_books['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import numpy as np

# Assuming 'data_array' is your NumPy array containing the data
# Reduce rows to half
half_vector = vector[:len(vector)//2, :]
```

```python
half_vec2 = half_vector[:len(vector)//2, :]
```

```python
half_vec2 = half_vec2[:len(vector)//4, :]
half_vec2.shape
```

```python
sliced_books = new_books.iloc[:len(new_books)//4, :]
```

```python
sliced_books.shape
```

```python
similary = cosine_similarity(half_vec2)
```

```python
similary.shape
```

```python
# Assuming 'new_books' is your DataFrame containing the data
index_of_first_match = new_books.index[new_books['book_title'] == 'Devil May Cry']

if not index_of_first_match.empty:
    print("Index of 'Harry Potter' row:", index_of_first_match[0])
else:
    print("No rows with 'book_title' equal to 'Harry Potter' found.")
```

```python
def recommend(book):
  index_of_first_match = new_books.index[new_books['book_title'] == book]

  if not index_of_first_match.empty:
    index = index_of_first_match[0]
    distances = sorted(list(enumerate(similary[index])),reverse=True , key = lambda x: x[1])
    for i in distances[1:6]:
      print(sliced_books.iloc[i[0]].book_title)
```

```python
recommend('The Claiming of Sleeping Beauty')
```
